here is an example of the different parametes and arguments sent to `plot` and to `scatter` functions.
basically `plot` is better if you need to plot x-y relation where all markers will be identical in size, color and identity
**but what happens when we want to change the color/marker of a point based on if it satisifed some given condition?**

then, we better use `scatter`

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data) # c=colour, s=size
plt.xlabel('entry a')
plt.ylabel('entry b')
```

```python
a=np.arange(0,20)
## in the plot function you supply x, later a y and then a format
# a format is a mixture of the desired display of the point, comprised of [color][marker][linestyle]
# for example: 'bo' == blue dots, 'g^' == green triangles, 'rs' == red squares

plt.plot(a,np.random.randn(20),'bo')
```

now we will do the same concept but using the `scatter` functions and mind which arguments we use **and how**:

```python
a=np.arange(0,20)
# in the `scatter` function we supply x,y as usual but we divide the display option of every point.
# basically, with providing valid list/array input - we can do that manually for each point.
# in this example, we change the color of the points based on a condition - whether or not the x value is larger then 10

plt.scatter(a,np.random.randn(20),c=a>10,marker='o')
```

Unfortunately, matplotlib has yet to simplify a way to change the marker style based on a conditon (like it allows you to do with color), but instead it expects to iterate over the data with each marker group at a time.

i'll try to explain it in the following example:

```python
markers=['o','*']
markers

#creating the x-y arrays
np.random.seed(42)
a=np.arange(20)
b=np.random.randn(20)
#creating the conditional, for now let's label it randomly with (0,1):
c=np.random.random_integers(0,1,20)

for k,m in enumerate(markers):

    i=(c==k) # creates a boolean array corresponding to whether or not current marker should be plotted
    
    plt.scatter(a[i], b[i],c=(a[i]%2==0),marker=m)
    plt.xticks(range(len(a)))
    
```

 ## Barplots

```python
categories = ['Meat','Dairy','Veggies','Sweets','legumes','Fruits','Carbs']
```

```python
amount = np.zeros(len(categories))
for i in range(len(categories)):
    amount[i] = np.random.randint(0,20)
```

```python
table = pd.DataFrame({"Categories":categories,"Amount":amount})
```

```python
table
```

```python
sns.barplot(x=table.Categories,y=table.Amount)
plt.xticks(table.Categories.index,np.arange(7))
for i in range(table.shape[0]):
    plt.text(x=i,y=table.loc[i,"Amount"],s=round(table.loc[i,"Amount"]/sum(table.Amount)*100,2),ha='center')
```

## Histograms

1. one ability very important to acquire in early stages of histogram plotting is to standardize the histogram.
i.e instead of having the y-axis showing the **count** of occurences of some values range,
it will show the frequency (or fraction out of total instances)

```python
random_hist = np.random.normal(size=1000)
```

```python
plt.hist(random_hist)
plt.title("just another normal distribution histogram")
plt.ylabel("Count",fontsize=14);
```

**note that the y-axis is describing the count**,<br>now let's see how to standardize it.

this is just so easy, there is an argument that can be set to `True` and it normalized the histogram:
`normed=True`

```python
plt.hist(random_hist,normed=True);
plt.tick_params(axis='x',colors='red')
```
